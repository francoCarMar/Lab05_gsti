```python
#4. UMAP para visualización de datos de alta dimensión
import umap
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Cargar el dataset de dígitos
digits = load_digits()
X = digits.data
y = digits.target

# Aplicar UMAP para reducir a 2 dimensiones
umap_model = umap.UMAP(n_components=2)
X_umap = umap_model.fit_transform(X)

# Graficar los datos reducidos
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis')
plt.title('Reducción de Dimensionalidad con UMAP')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()

```
