```python
# 5. Reducción de Dimensionalidad con PCA para una base de datos grande

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
import numpy as np

# 1. Cargar el dataset MNIST 
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 2. Combinar train y test en un solo conjunto
X = np.concatenate([X_train, X_test], axis=0)
y = np.concatenate([y_train, y_test], axis=0)

# 3. Aplanar las imágenes (de 28x28 a 784)
X = X.reshape(-1, 28*28)

# 4. Normalizar los valores (0-1)
X = X / 255.0

# Aplicar PCA para reducir a 50 dimensiones
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X)

# 6. Graficar la varianza explicada por los primeros 50 componentes
plt.figure(figsize=(8,5))
plt.plot(range(1, 51), pca.explained_variance_ratio_[:50], marker='o', color='blue')
plt.title('Varianza Explicada por los Primeros 50 Componentes Principales')
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 7. (Opcional) Mostrar cuánta varianza total explican los 50 componentes
print(f"Varianza total explicada por los 50 componentes: {pca.explained_variance_ratio_.sum():.2%}")
```
