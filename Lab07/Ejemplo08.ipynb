```python
# 8. PCA para compresin de imgenes
from sklearn.decomposition import PCA
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt

# Cargar una imagen de ejemplo
china = load_sample_image("china.jpg")

# Aplanar la imagen para aplicar PCA
china_flat = china.reshape(-1, 3)

# Aplicar PCA para reducir la dimensionalidad
pca = PCA(n_components=3) # Changed n_components from 50 to 3
china_pca = pca.fit_transform(china_flat)

# Reconstruir la imagen a partir de las componentes principales
china_reconstructed = pca.inverse_transform(china_pca).reshape(china.shape)

# Graficar la imagen original y la reconstruida
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(china)
ax[0].set_title('Imagen Original')
ax[1].imshow(china_reconstructed)
ax[1].set_title('Imagen Reconstruida con PCA')
plt.show()
```
